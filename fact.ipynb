```python
num= int(input("Enter any number:"))

if(num<0):
  print("Factorial not possible.")
elif num==0:
  print("Factorial is 1")
else:
  fact= 1;
  for x in range(1,num+1):
    fact=fact*x
  print("Factorial is",fact)
```
