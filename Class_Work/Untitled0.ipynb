```python
#to find out hcf
num1 = int(input("enter first number"))
num2= int(input("enter second number"))
if num2==0:
  print("hcf",num1)
else:
  while(num2!=0):
    rem=num1%num2
    num1=num2
    num2=rem
#--------------
print("HCF",num1)



```
